```python
import pandas as pd
import matplotlib
#matplotlib.use('Agg')
import numpy as np
#import seaborn as sns
from matplotlib import pyplot as plt
```

```python
systsecm='CONTROL'
rmsd_dir="/Users/fatima/Stanford/70S/Trajectories/%s/onionRMSD/col2/" % systsecm
out_dir="/Users/fatima/Stanford/WritingSecM/ONION_RMSD/"
```

```python
rmslist=['rmsd10A.dat','rmsd20A.dat','rmsd30A.dat','rmsd40A.dat','rmsd50A.dat','rmsd60A.dat',
         'rmsd70A.dat','rmsd80A.dat','rmsd200A.dat']
```

```python
rmslistnames=['rmsd10A','rmsd20A','rmsd30A','rmsd40A','rmsd50A','rmsd60A',
         'rmsd70A','rmsd80A','rmsd200A']
```

```python

rmsd_plot=[]
for rmsfile in rmslist:
    rmsd_plot.append(np.loadtxt(rmsd_dir+rmsfile))
    
df = pd.DataFrame(rmsd_plot).transpose()

df.columns=rmslistnames

print(df)

fig = plt.figure()
ax = fig.add_subplot(111)

plt.xlabel("Frames", fontsize=16)
plt.ylabel("RMSD (A)", fontsize=16)
plt.title("%s" % systsecm, fontsize=16)
xmin, xmax, ymin, ymax = plt.axis([-10, 260, 0, 18])

#handles, labels = plt.gca().get_legend_handles_labels()
#print(handles)

#print(df['rmsd10A'])

for item in rmslistnames:
    plt.plot(df[item])


# plt.axvline(x=50,ls='--')
# plt.axvline(x=96,ls='--')
# plt.axvline(x=146,ls='--')
# plt.axvline(x=197,ls='--')

plt.axvline(x=50,ls='--')
plt.axvline(x=101,ls='--')
plt.axvline(x=152,ls='--')
plt.axvline(x=203,ls='--')

#ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
#plt.legend()
plt.savefig(out_dir+systsecm+'_onion_rmsd.pdf')
    
```
